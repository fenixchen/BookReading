# matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x = np.arange(1, 20, 0.1)
y = np.sin(x)
plt.plot(x, y)
plt.show
```

```python
x = np.arange(1, 20, 0.1)
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x, y1, label='sin')
plt.plot(x, y2, linestyle='--', label='cos')
plt.xlabel("x")
plt.ylabel("y")
plt.title("Test SIN and COS")
plt.legend()
plt.show()
```
